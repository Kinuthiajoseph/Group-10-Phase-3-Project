***Train-Test Split with Scikit-Learn***

N\B: It's a good practice to split your dataset into training and testing sets before performing any preprocessing, especially when dealing with modeling tasks. This helps prevent data leakage, which occurs when information from the testing set inadvertently influences the training process, leading to overly optimistic performance metrics.

Here is our approach: 

* Data Splitting: Divide the dataset into two separate sets: one for training and one for testing.

* Preprocessing: Perform data preprocessing steps (cleaning, feature engineering, transformation, encoding, etc.) on the training set. 

* Modeling and Evaluation: Train and validate your models using the preprocessed training set.

##### Correlation Heatmap for all Numerical Columns

***
Here are the observations :

* Many features in the dataset show perfect correlation, such as "Total day charge" and "Total day minutes," as well as the same pattern for other charge and minutes features.
* This high correlation is expected because charges are directly influenced by the corresponding minutes used.
* A correlation coefficient of 1 indicates perfect correlation, implying these features move together linearly and can be expressed as combinations of each other.
* Perfect correlation can lead to challenges in linear models like unstable coefficient estimates and difficulties in interpreting results, due to violating the independence assumption.
* However, in nonlinear models like decision trees and random forests, the impact of perfect multicollinearity is less pronounced since these models capture complex relationships without the same linear independence assumptions.
* Nonlinear models may handle correlated features better, but it's important to be cautious of potential effects on feature importance scores.
* In these models, one of the correlated features might be prioritized over others in terms of importance due to their shared information.
